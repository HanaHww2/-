# Lab-02-1 linear regression

````python
import tensorflow as tf
tf.set_random_seed(777)  # for reprducibilty, 씨드 설정으로 랜덤값 재현가능하도록
````

## X and Y data

````python
x_train = [1, 2, 3]
y_train = [1, 2, 3]
````

## Variable
- 텐서플로가 활용하는 변수
- trainable variable

````python
# Try to find value for W and b to compute y_data = x_data * W + b  
# We know that W should be 1 and b should be 0
# But let's TensorFlow figure it out 
# tf.random_normal([1]) 차원이 1인 랜덤값

W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
````

## Our Model

````python
# Out hypothesis XW+b
hypothesis = x_train * W + b

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
````

## Minimize - GradientDescent

````python
# graph(node) for minimizing

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
````

## Prepare session

````python
# Launch the graph in a session.
sess = tf.Session()

# Initializes global variables in the graph.
sess.run(tf.global_variables_initializer())
````

## Fit the line

````python
for step in range(2001):
    sess.run(train)
    if step % 200 == 0: # if step % 20 == 0:, 프린트하며 확인할 횟수 조절
        print(step, sess.run(cost), sess.run(W), sess.run(b))
````

# Lab-02-2 linear regression feed

## Variable

````python
# Try to find value for W and b to compute y_data = x_data * W + b  
# We know that W should be 1 and b should be 0
# But let's TensorFlow figure it out 
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
````

## X and Y data

````python
#### Now we can use X and Y in place of x_data and y_data
#### placeholders for a tensor that will be always fed using feed_dict
#### See http://stackoverflow.com/questions/36693740/

X = tf.placeholder(tf.float32)
# X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])
````

## Our Model

````python
# Out hypothesis XW+b
hypothesis = X * W + b

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))
````

## Minimize

````python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
````

## Prepare session

````python
# Launch the graph in a session.
sess = tf.Session()
# Initializes global variables in the graph.
sess.run(tf.global_variables_initializer())
````

## Fit the line

````python
for step in range(2001):
    cost_val, W_val, b_val, _ = \
        sess.run([cost, W, b, train], # 리스트로 한꺼번에 메소드 수행 가능
                 feed_dict={X: [1, 2, 3], Y: [1, 2, 3]})
    if step % 200 == 0:
        print(step, cost_val, W_val, b_val)
````

#### Learns best fit W:[ 1.],  b:[ 0]

## Testing our model

````python
print(sess.run(hypothesis, feed_dict={X: [5]}))
print(sess.run(hypothesis, feed_dict={X: [2.5]}))
print(sess.run(hypothesis, feed_dict={X: [1.5, 3.5]}))
````

## Fit the line with new training data

````python
for step in range(2001):
    cost_val, W_val, b_val, _ = \
        sess.run([cost, W, b, train],
                 feed_dict={X: [1, 2, 3, 4, 5],
                            Y: [2.1, 3.1, 4.1, 5.1, 6.1]})
    if step % 200 == 0:
        print(step, cost_val, W_val, b_val)
````

## Testing our model

````python
print(sess.run(hypothesis, feed_dict={X: [5]}))
print(sess.run(hypothesis, feed_dict={X: [2.5]}))
print(sess.run(hypothesis, feed_dict={X: [1.5, 3.5]}))
````

# Lab-02-3 linear regression tensorflow.org

````python
# From https://www.tensorflow.org/get_started/get_started
# import tensorflow as tf
````

## Variable

````python
# Model parameters
W = tf.Variable([.3], tf.float32)
b = tf.Variable([-.3], tf.float32)

# Model input and output
x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
````

## Our Model

````python
linear_model = x * W + b

# cost/loss function
loss = tf.reduce_sum(tf.square(linear_model - y))  # sum of the squares
````

## Minimize

````python
# optimizer
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(loss)
````

## X and Y data

````python
# training data
x_train = [1, 2, 3, 4]
y_train = [0, -1, -2, -3]
````

## Fit the line

````python
# training loop
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)  # reset values to wrong
for i in range(1000):
    sess.run(train, {x: x_train, y: y_train})
````

## evaluate training accuracy

````python
# evaluate training accuracy
curr_W, curr_b, curr_loss = sess.run([W, b, loss], {x: x_train, y: y_train})
print("W: %s b: %s loss: %s" % (curr_W, curr_b, curr_loss))
````
